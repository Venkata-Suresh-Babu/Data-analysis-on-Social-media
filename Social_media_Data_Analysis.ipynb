# **Import the data analysis Python pacakes (pandas & matlplot lib)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
time_wasters = pd.read_csv('Time-Wasters on Social Media.csv')
time_wasters
```

```python
#find Null values
time_wasters.isnull().sum()
```

```python
time_wasters.dropna(inplace = True)
time_wasters
```

```python
time_wasters.describe()
```

```python
#getting column names in  dataframe
time_wasters.columns
```

```python
#list the professions
time_wasters['Profession'].value_counts()
```

```python
#Gender count
time_wasters['Gender'].value_counts()
```

```python
age_counts = time_wasters['Age'].value_counts().sort_values(ascending=False)
# Create a bar graph
plt.figure(figsize=(20, 5))
age_counts.plot(kind='bar', color="blue")
# Add count labels above bars
for i, count in enumerate(age_counts):
    plt.text(i, count, str(count), fontsize=10, ha='center', va='bottom')
# Customize the plot
plt.legend()
plt.title("Social media presence by age (Descending)")
plt.xlabel("Age")
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.show()
```

```python
#ploting a bar graph on genders
gender = time_wasters['Gender'].value_counts()
gender_count = gender.plot(kind='bar', figsize=(5, 5), color = "blue")
for i, count in enumerate(gender):
  gender_count.text(i, count, str(count), fontsize=10, ha='center', va='bottom')
plt.legend()
plt.title("Gender Count")
plt.xlabel("Genders")
plt.ylabel('Count')
plt.show()
```

```python
#Demographics
Demographics = time_wasters['Demographics'].value_counts()
Demographics_count = Demographics.plot(kind='bar', figsize=(5, 8), color = "orange")
for i, count in enumerate(Demographics):
  Demographics_count.text(i, count, str(count), fontsize=10, ha='center', va='bottom')
plt.legend()
plt.title("Demographics Count")
plt.xlabel("Location")
plt.ylabel('Count')
plt.show()
```

```python
#Profession Count
profession_counts = time_wasters['Profession'].value_counts()
pro_count = profession_counts.plot(kind='area', figsize=(15, 5), color="green")
# Adding data labels
for i, count in enumerate(profession_counts):
    pro_count.text(i, count, str(count), fontsize=12, ha='center', va='bottom')
plt.legend()
plt.title("Social media presence by profession")
plt.xlabel("Profession")
plt.ylabel("Count")

plt.show()
```

```python
addition_time = time_wasters['Addiction Level'].value_counts()
addition_time
```

```python
# Group the data by gender and addiction level
grouped_data = time_wasters.groupby(['Gender', 'Addiction Level'])['Addiction Level'].value_counts()

# Create subplots for each gender
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 6))

# Iterate through each gender and plot the corresponding counts
for i, gender in enumerate(['Male', 'Female', 'Other']):
    gender_data = grouped_data.loc[gender]
    gender_data.plot(kind='bar', ax=axes[i], title=f"{gender}")
    axes[i].set_xlabel('Addiction Level')
    axes[i].set_ylabel('Count')

    # Annotate data values on each bar
    for p in axes[i].patches:
        axes[i].annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom')

# Add a title to the entire visualization
plt.suptitle('Addiction Level Counts by Gender')

# Create a legend with the specified title and position
plt.legend(title='Addiction Level Interpretation', labels=['Addiction level: 0[Low], 7[High]'], loc='upper right')

plt.tight_layout()
plt.show()
```

```python
male_students = time_wasters[time_wasters['Gender'] == 'Male']
female_students = time_wasters[time_wasters['Gender'] == 'Female']
```

```python
males = male_students.groupby('Platform').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))

# Add data labels
for i, v in enumerate(male_students.groupby('Platform').size()):
    males.text(v + 1, i, str(v), va='center')

# Set the title, labels, and spines
plt.title('Male Students by Platform')
plt.xlabel('Count')
plt.ylabel('Platform')
plt.gca().spines[['top', 'right',]].set_visible(False)

plt.show()
```

```python
females = male_students.groupby('Platform').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))

# Add data labels
for i, v in enumerate(female_students.groupby('Platform').size()):
    females.text(v + 1, i, str(v), va='center')

# Set the title, labels, and spines
plt.title('Female Students by Platform')
plt.xlabel('Count')
plt.ylabel('Platform')
plt.gca().spines[['top', 'right',]].set_visible(False)

plt.show()
```

```python
time_wasters['CurrentActivity'].value_counts()
```

```python
watch_count = time_wasters['Watch Time'].value_counts()
time_plot = watch_count.plot(kind = "bar", figsize = (10, 7), color = "maroon")
for i, count in enumerate(watch_count):
  time_plot.text(i, count, str(count), fontsize=15, ha='center', va='bottom', fontweight = 'bold', color = 'green')
plt.title("Most Watch Time")
plt.xlabel("Watch Time")
plt.ylabel('Count')
plt.show()
```

```python
category_count = time_wasters['Video Category'].value_counts()
category_plot = category_count.plot(kind = "bar", figsize = (14, 8), color = "silver")
for i, count in enumerate(category_count):
  category_plot.text(i, count, str(count), fontsize=15, ha='center', va='bottom', fontweight = 'bold', color = 'green')
plt.title("Content most watched")
plt.xlabel("Content")
plt.ylabel('Count')
plt.show()
```

```python

# Analyze the impact of social media usage on user productivity
average_productivity_loss = time_wasters['ProductivityLoss'].mean()

# Plot the average productivity loss
plt.figure(figsize=(6, 4))
plt.bar(x=['Average Productivity Loss'], height=[average_productivity_loss])
plt.xlabel('Productivity Loss')
plt.ylabel('Hours')
plt.title('Average Productivity Loss due to Social Media Usage')
plt.show()
```

```python
correlation = time_wasters['Self Control'].corr(time_wasters['Addiction Level'])

# Plot the correlation between self-control and addiction levels
plt.figure(figsize=(8, 6))
sns.heatmap(time_wasters[['Self Control', 'Addiction Level']].corr(), annot=True)
plt.title('Correlation between Self Control and Addiction Level')
plt.show()
```

```python
device_type_counts = time_wasters['DeviceType'].value_counts()
# Analyze OS impact on Platform choices
os_platform_crosstab = pd.crosstab(time_wasters['OS'], time_wasters['Platform'])

# Plot OS impact on Platform choices (Here, you can choose between bar chart or stacked bar chart)
plt.figure(figsize=(10, 6))
os_platform_crosstab.plot(kind='bar', stacked=False, color=['skyblue', 'lightgreen', 'gold'])
plt.xlabel('Operating System')
plt.ylabel('Number of Users')
plt.title('Platform Usage Distribution by Operating System')
plt.legend(title='Platform', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
watch_reason_counts = time_wasters['Watch Reason'].value_counts()

# Plot reasons why users watch videos on social media
plt.figure(figsize=(8, 6))
watch_reason_counts.plot(kind='bar', color=['skyblue', 'lightgreen', 'gold', 'coral'])
plt.xlabel('Watch Reason')
plt.ylabel('Number of Users')
plt.title('Distribution of Users by Watch Reason')
plt.show()
```

```python
# Analyze user satisfaction with different platforms
platform_satisfaction_avg = time_wasters.groupby('Platform')['Satisfaction'].mean()

# Plot user satisfaction with different platforms
plt.figure(figsize=(8, 6))
plt.bar(platform_satisfaction_avg.index, platform_satisfaction_avg.values)
plt.xlabel('Platform')
plt.ylabel('Average Satisfaction Score')
plt.title('Average User Satisfaction by Platform')
plt.show()
```
